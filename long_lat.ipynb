```python
import os
import random
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
# Generating random coordinates
def generate_random_coordinates(min_lat=24.0, max_lat=39.8, min_lon=44.0, max_lon=63.3):
    latitude = round(random.uniform(min_lat, max_lat), 2)
    longitude = round(random.uniform(min_lon, max_lon), 2)
    return latitude, longitude


# Generating a dataframe including coordinates
def coordinate_df(min_lat=24.0, max_lat=39.8, min_lon=44.0, max_lon=63.3, num_points=10):
    coordinate_dict = {'Latitude': [], 'Longitude': []}

    for point in range(num_points):
        lat, lon = generate_random_coordinates(min_lat, max_lat, min_lon, max_lon)
        coordinate_dict['Latitude'].append(lat)
        coordinate_dict['Longitude'].append(lon)

    df_coor = pd.DataFrame(coordinate_dict)
    return df_coor
    return df_coor
```

```python
# Save DataFrame as an Excel file
def save_df_excel(name, df, path=None):
    if not name.endswith('.xlsx'):
        name += '.xlsx'

    if path is None:
        path = ''
    else:
        path = os.path.normpath(path)  
    # Ensure the path exists; create it if necessary
    if not os.path.exists(path):
        os.makedirs(path)

    file_path = os.path.join(path, name)
    df.to_excel(file_path, index=False)
```

```python

# Function to read Excel and create a DataFrame
def df_creator(excel_file):
    df = pd.read_excel(excel_file)
    return df

# Function to show coordinates on a map
def show_on_map(excel_file='coordinates.xlsx'):
    df = df_creator(excel_file)

    center_lat, center_lon = df.iloc[0]['Latitude'], df.iloc[0]['Longitude']
    map_center = [center_lat, center_lon]
    mymap = folium.Map(location=map_center, zoom_start=6)

    marker_cluster = MarkerCluster().add_to(mymap)

    for index, row in df.iterrows():
        lat, lon = row['Latitude'], row['Longitude']
        popup_text = f"Latitude: {lat}, Longitude: {lon}"
        folium.Marker(location=[lat, lon], popup=popup_text).add_to(marker_cluster)

    return mymap


```

```python
show_on_map(excel_file='50_points.xlsx')
```

```python

import sys
from PyQt6.QtWidgets import QWidget, QApplication, QVBoxLayout
from PyQt6.QtWidgets import QApplication
from PyQt6.QtWebEngineWidgets import QWebEngineView


class Example(QWidget):

    def __init__(self):
        super().__init__()

        self.initUI()

    def initUI(self):

        vbox = QVBoxLayout(self)

        self.webEngineView = QWebEngineView()
        self.loadPage()

        vbox.addWidget(self.webEngineView)

        self.setLayout(vbox)

        self.setGeometry(300, 300, 350, 250)
        self.setWindowTitle('QWebEngineView')
        self.show()

    def loadPage(self):

        with open('test.html', 'r') as f:

            html = f.read()
            self.webEngineView.setHtml(html)

def main():

    app = QApplication(sys.argv)
    ex = Example()
    sys.exit(app.exec_())


if __name__ == '__main__':
    main()
```
